```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
```

This notebook combines the forecast for the available quantum volume with the formulas for the size of QV and QAE circuits and the noise model into an estimate of the Fisher information per second that can be generated.

```python
N = 2000
```

```python
def tim(T):
    """
    Time in months since november 2018
    """
    t = 12*(T-((2019-2/12)))
    return t

def q(T):
    """
    Prediction for the number of qubits that can pass a QV experiment.
    """
    return 2 + (T/6)
    
```

```python
#Plot of the number of available QV qubits
fig,ax = plt.subplots()
X = [(2019-2/12) + 20 * (s+0.5)/N for s in range(N)]
Y = [ q(tim(T)) for T in X ]
ax.plot(X,Y, color='black')
#plt.legend("available qubits")
#plt.savefig("qT.png")
plt.show()
```

```python
#The essential parameters of the models
aQ = 0.48
a2 = 0.24
b2 = 8.13
a1 = 17.35
b1 = 0.46
R = 0
Q = 8
```

```python
def c(T):
    """
    The probability to execute a slice of a QAE circuit of 100 gates without an error at time T.
    """
    return aQ**(100 / (a2*(q(T)**3)+b2*(q(T)**2)))

def a(T):
    """
    The probability to execute a QAE circuit with Q qubits and R amplification rounds without an error at time T.
    Here it is assumed that b=1, so no measurement or initialization errors occur.
    """
    return c(T)**( ( a1*(2*R+1) * (10**(b1*Q)) ) / ( 100 ) )
```

```python
for T in [year for year in range(2020, 2080, 8)]:
    print(T, q(tim(T)), c(tim(T)), a(tim(T)))
```

```python
#Plot of the error parameters.
fig,ax = plt.subplots()
X = [(2019-2/12) + 60 * (s+0.5)/N for s in range(N)]
Y = [ c(tim(T)) for T in X ]
ax.plot(X,Y, color='blue')
Y2 = [ a(tim(T)) for T in X ]
ax.plot(X,Y2, color='green')
plt.legend(['$c_{T}$', '$a_{T}$'])
ax.set_xlabel('Year')
ax.set_ylabel('Probability')
#plt.savefig("aT_cT.png")
plt.show()
```

```python
def R_def(T):
    """
    The number of amplification rounds that optimizes the Fisher information.
    """
    ct = c(T) ** ( ( a1 * (10**(b1*Q)) ) / ( 100 ) )
    y = -(1/2) + ( 1 / ( 1-(ct*ct) ) )
    return int(y)

def aOp(T):
    """
    The optimal operational probability to run a cricuit without errors.
    """
    return c(T)**( ( a1*(2*R_def(T)+1) * (10**(b1*Q)) ) / ( 100 ) )

for year in range(2200,2250, 5):
    print(year, R_def(tim(year)), int(R_def(tim(year))), (1+2*int(R_def(tim(year))))*((c(tim(year))**((7*a1*(10**(b1*Q)))/(100*Q)))**(1+2*int(R_def(tim(year))))) , c(tim(year))**((7*a1*(10**(b1*Q)))/(100*Q)) )
```

```python
fig,ax = plt.subplots()
X = [(2019-2/12) + 60 * (s+0.5)/N for s in range(N)]
Y = [ R_def(tim(T)) for T in X ]
ax.plot(X,Y, color='magenta')
#Y2 = [ aOp(tim(T)) for T in X ]
#ax.plot(X,Y2, color='darkgreen')
Y3 = [ (1+2*R_def(tim(T)))*aOp(tim(T)) for T in X ]
ax.plot(X,Y3, color='darkblue')
plt.legend(['$R$', '$(2R+1)a_{Op}$'])
ax.set_xlabel('Year')
ax.set_ylabel('Optimal number of amplification rounds')
#plt.savefig("RT.png")
plt.show()
```

```python
fig,ax = plt.subplots()
X = [(2019-2/12) + 60 * (s+0.5)/N for s in range(N)]
Y = [ aOp(tim(T)) for T in X ]
ax.plot(X,Y, color='darkgreen')
ax.set_xlabel('Year')
ax.set_ylabel('Probability')
#plt.savefig("RT.png")
plt.show()
```

```python
def Ips(T):
    """
    The Fisher information per second.
    """
    return 4 * (10**6) * (aOp(T)) * (2*R_def(T)+1) / ( a1 * (10**(b1*Q)) )

for year in [2023, 2050, 2100, 2150, 2200, 2250, 2300]:
    print(year, Ips(tim(year)))

print(Q,R)
```

```python
fig,ax = plt.subplots()
X = [(2019-2/12) + 60 * (s+0.5)/N for s in range(N)]
Y = [ Ips(tim(T)) for T in X ]
ax.plot(X,Y, color='red')
ax.set_xlabel('Year')
ax.set_ylabel('Fisher information per second')
plt.legend(['$\mathcal{I}\,(1/2)$'])
#plt.savefig("FI.png")
plt.show()
```

### Classical MCI

```python
%%time
def ef(x):
    mu = 1
    sigma = 2
    labda = 7
    return labda*np.exp(-x*labda)*np.exp(-0.5*(((x-mu)/sigma)**2)) / np.sqrt(2*np.pi*(sigma**2))

N=100000#0
z = 0.0
for i in range(N):
    x = 0.15 + (2*np.sqrt(2)*np.pi-0.15) * (i+0.5)/N
    z += ef(x) * (2*np.sqrt(2)*np.pi-0.15)/N
print(z)
```

## Shallow functions

Below we repeat the steps above for a shallow function implementation.

```python
"""
aQ = 0.48
a2 = 0.24
b2 = 8.13
a1 = 17.35
b1 = 0.46
R = 0
Q = 8
"""
a1 = 6.0
b1 = 0.10
a3 = 1.0
b3 = 3.0
R = 0
Q = 27
```

```python
def c(T):
    return aQ**(100 / (a2*(q(T)**3)+b2*(q(T)**2)))

def a(T):
    #return c(T)**( 7 * ( a1*(2*R+1) * (10**(b1*Q)) ) / ( 100 * Q ) )
    return c(T)**( ( a3*(2*R+1) * (Q**b3) ) / ( 100 ) )
```

```python
fig,ax = plt.subplots()
X = [(2019-2/12) + 60 * (s+0.5)/N for s in range(N)]
Y = [ c(tim(T)) for T in X ]
ax.plot(X,Y, color='blue')
Y2 = [ a(tim(T)) for T in X ]
ax.plot(X,Y2, color='green')
plt.legend(['$c_{T}$', '$a_{T}$'])
ax.set_xlabel('Year')
ax.set_ylabel('Probability')
#plt.savefig("aT_cT_shallow.png")
plt.show()
```

```python
def R_def(T):
    ct = c(T) ** ( ( a3 * (Q**(b3)) ) / ( 100 ) )
    y = -(1/2) + ( 1 / ( 1-(ct*ct) ) )
    return int(y)

def aOp(T):
    return c(T)**( ( a3*(2*R_def(T)+1) * (Q**(b3)) ) / ( 100 ) )
```

```python
fig,ax = plt.subplots()
X = [(2019-2/12) + 60 * (s+0.5)/N for s in range(N)]
Y = [ R_def(tim(T)) for T in X ]
ax.plot(X,Y, color='magenta')
#Y2 = [ aOp(tim(T)) for T in X ]
#ax.plot(X,Y2, color='darkgreen')
Y3 = [ (1+2*R_def(tim(T)))*aOp(tim(T)) for T in X ]
ax.plot(X,Y3, color='darkblue')
plt.legend(['$R$', '$(2R+1)a_{Op}$'])
ax.set_xlabel('Year')
ax.set_ylabel('Optimal number of amplification rounds')
#plt.savefig("RT_shallow.png")
plt.show()
```

```python
def Ips(T):
    return 4 * (10**6) * (aOp(T)) * (2*R_def(T)+1) / ( a3 * (Q**(b3)) )

for year in [2023, 2050, 2075, 2100]:
    print(year, Ips(tim(year)))

print(Q,R)
```

```python
fig,ax = plt.subplots()
X = [(2019-2/12) + 60 * (s+0.5)/N for s in range(N)]
Y = [ Ips(tim(T)) for T in X ]
ax.plot(X,Y, color='red')
ax.set_xlabel('Year')
ax.set_ylabel('Fisher information per second')
plt.legend(['$\mathcal{I}\,(1/2)$'])
#plt.savefig("FI_shallow.png")
plt.show()
```
